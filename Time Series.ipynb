```python
data["Close"].rolling(window=5).mean().head(20) # 5 days moving avareage
```

```python
sma_21 = data["Close"].rolling(window=21).mean() # 21 days simple moving avareage
```

```python
sma_55 = data["Close"].rolling(window = 55).mean() # 55 days simple moving average
```

```python
sma_100 = data["Close"].rolling(window = 100).mean() # 100 days simple moving average
```

```python
sma_200 = data["Close"].rolling(window = 200).mean() # 100 days simple moving average
```

```python

sns.lineplot(x = data.index, y = sma_21 , legend= True, palette=["red"])
sns.lineplot(x = data.index, y = sma_55, legend= True)
sns.lineplot(x = data.index, y = sma_100, legend= True)
sns.lineplot(x = data.index, y = sma_200, legend= True)
sns.lineplot(x = data.index, y = data["Close"], legend= True)

plt.legend(["21 EMA", "55 EMA", "100 EMA","200 EMA", "Price"])
plt.xlim((datetime(2020, 1, 1) , datetime(2020, 3, 31)))
plt.xticks(rotation = 45, ha = "right")
plt.show()
```
