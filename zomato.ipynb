### Group By

```python
final_data.columns
```

```python
gb = final_data.groupby(["Aggregate rating", "Rating color", "Rating text"]).
```

```python
gb.size()
```

```python
ratings = gb.size().reset_index() # to convert into dataframe
```

```python
ratings
```

```python
ratings.rename(columns={0:"Rating Count"}, inplace=True) # rename dataframe comlumn name
```

```python
ratings
```

## observations:
1. Whenever the rating is from 4.9 to 4.5 this indicate - Excellent 
                               4.0 to 4.4               - Very Good
                               3.5 to 3.9               - Good
                               2.5 to 3.4               - Average
                               1.8 to 2.5               - Poor
                               0.0                      - Not Rated

2. The maximum people has not rated

```python
ratings.sample(5)
```

```python
sns.barplot(data = ratings, x = "Aggregate rating", y = "Rating Count")
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(data = ratings, x = "Aggregate rating", y = "Rating Count")
plt.xticks(rotation = 45, ha = "center") # left, right , center
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(data = ratings, x = "Aggregate rating", y = "Rating Count", hue = "Rating color" )
plt.xticks(rotation = 45, ha = "center") # left, right , center
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(data = ratings, x = "Aggregate rating", y = "Rating Count", hue = "Rating color", palette = ["white", "red", "orange", "yellow", "#90EE90", "#013220"] )
plt.xticks(rotation = 45, ha = "center") # left, right , center
plt.show()
```

```python
# Prevent bars from becoming thin: (dodge parameter)
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(data = ratings, x = "Aggregate rating", y = "Rating Count", hue = "Rating color", palette = ["#D3D3D3", "red", "orange", "yellow", "#90EE90", "#013220"], dodge = False )
plt.xticks(rotation = 45, ha = "center") # left, right , center
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(data = ratings, x = "Aggregate rating", y = "Rating Count", hue = "Rating text", palette = ["#D3D3D3", "red", "orange", "yellow", "#90EE90", "#013220"], dodge = False )
plt.xticks(rotation = 45, ha = "center") # left, right , center
plt.show()
```
